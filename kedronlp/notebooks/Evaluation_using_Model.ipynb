## Evaluation using an AI Model

In the following approach to create question answer pairs, an AI Model, in this case Google Gemini Advanced, is given a context which is the abstract of a paper from the PubMed Dataset and instructed to create a question of a specific type (for e.g. a Yes / No question) out of that context.  

Using this approach, the context provided to the AI Model becomes the answer and the question generated by the independent AI Model becomes the question for out evaluation dataset. This can then be tested on our LLM Model to see how well it performs on the generated question answer pairs using metrics like BLEU, BLEURT, etc.
